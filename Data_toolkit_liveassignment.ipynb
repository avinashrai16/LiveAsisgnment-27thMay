Q1. Demonstrate three different methods for creating identical 2D arrays in

```python
import numpy as np
```

```python
 #Using copy method:

# Create the first 2D array
array1 = np.array([[1, 2, 3], [4, 5, 6]])

# Create an identical copy of the array
array2 = array1.copy()

print("Array 1:")
print(array1)
print("-----------------")
print("Array 2:")
print(array2)
```

```python
# Using simple assignment (both variables will reference the same array):

# Create the first 2D array
array1 = np.array([[1, 2, 3], [4, 5, 6]])

# Create another reference to the same array
array2 = array1

print("Array 1:")
print(array1)
print("-----------------")
print("Array 2:")
print(array2)
```

```python
# Using np.copy function:

# Create the first 2D array
array1 = np.array([[1, 2, 3], [4, 5, 6]])

# Create an identical copy of the array using np.copy
array2 = np.copy(array1)

print("Array 1:")
print(array1)
print("-----------------")
print("Array 2:")
print(array2)
```

Q3. Explain the following terms:
* The difference in np.array, np.asarray and np.asanyarray.

   ***np.array:*** Creates a new array from any object,
  <U>Always creates a new array, even if the input is already an ndarray.</U>
  
  ***np.asarray:*** Converts the input to an ndarray, <u>but does not create a new array if the input is already an ndarray.</U>

  ***np.asanyarray:*** Similar to *np.asarray*, but it passes through subclasses of ndarray. This is useful when working with objects like matrices or masked arrays, where you want to retain the subclass type.<u>Converts the input to an ndarray or a subclass, without creating a new array if the input is already an ndarray or subclass.</u>
---
* The difference between Deep copy and shallow copy.

  the terms "deep copy" and "shallow copy" refer to the ways in which objects are copied.

  ***Shallow Copy***:A shallow copy creates a new object, but it does not create copies of nested objects. Instead, it inserts references to the objects found in the original. <u>This means that changes to the nested objects in the copied object will reflect in the original object</u>.

  ***Deep Copy***:A deep copy creates a new object and recursively copies all objects found in the original. <u>This means that changes to the nested objects in the copied object will not affect the original object<u>.

Q14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset in a Pandas DataFrame
np.random.seed(0)  # for reproducibility
dates = pd.date_range(start='2024-01-01', end='2024-01-10')
temperature = np.random.randint(20, 30, size=len(dates))
humidity = np.random.randint(40, 60, size=len(dates))

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})
# df.head()

# Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots(figsize=(12, 8))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # creates a secondary y-axis (right y-axis)
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# # Set the title of the plot
plt.title('Temperature and Humidity Over Time')
plt.show()
```

Q15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

Q16 Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', label='Histogram')

# Compute the  probability density function ~ PDF of the normal distribution
x = np.linspace(-4, 4, 1000)
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Overlay a line plot representing the normal distribution's PDF
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

Q17
Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrant = np.zeros_like(x, dtype=str)
quadrant[(x > 0) & (y > 0)] = 'Quadrant I'
quadrant[(x < 0) & (y > 0)] = 'Quadrant II'
quadrant[(x < 0) & (y < 0)] = 'Quadrant III'
quadrant[(x > 0) & (y < 0)] = 'Quadrant IV'

# Create a DataFrame
data = {'x': x, 'y': y, 'Quadrant': quadrant}
df = pd.DataFrame(data)

# Create a scatter plot with Seaborn
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Add legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

Q18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'SineWave Function'.

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave function
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new Bokeh plot
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer for the sine wave
p.line(x, y, line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the plot
show(p)
```

Q19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', min_height=400, min_width=600)

# Map colors to the values
mapper = factor_cmap(field_name='Category', palette=Viridis256, factors=categories)

# Add bars to the plot
p.vbar(x='Category', top='Value', width=0.9, source=source, legend_field="Category",
       line_color='white', fill_color=mapper)

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Category", "@Category"),
    ("Value", "@Value")
]
p.add_tools(hover)

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

Q20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
